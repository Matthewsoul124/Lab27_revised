# Метод разделимых наименьших квадратов. Квантовые дефекты

Методом разделимых наименьших квадратов построим модель вида

$y$ = $\beta1$ - $\beta2$$x$ - $\frac{arctan(\frac{\beta3}{x-\beta4})}{\pi}$

для данных из исследований квантовых дефектов в атомах йода. Предиктор $x$ - возбуждённое состояние энергии, переменная отклика $y$ - число квантовых дефектов.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import linalg
from scipy.optimize import curve_fit
```

Запишем имеющиеся у нас данные:

```python
x = np.array([-4868.68, -4868.09, -4867.41, -3375.19, -3373.14, -3372.03, -2473.74, -2472.35, -2469.45, -1894.65, -1893.40, -1497.24, -1495.85, -1493.41, -1208.68, -1206.18, -1206.04, -997.92, -996.61, -996.31, -834.94, -834.66, -710.03, -530.16, -464.17 ])
y = np.array([0.252429, 0.252141, 0.251809, 0.297989, 0.296257, 0.295319, 0.339603, 0.337731, 0.333820, 0.389510, 0.386998, 0.438864, 0.434887, 0.427893, 0.471568, 0.461699, 0.461144, 0.513532, 0.506641, 0.505062, 0.535648, 0.533726, 0.568064, 0.612886, 0.624169])
```

Теперь поработаем с функцией $y$ = $\beta1$ - $\beta2$$x$ - $\frac{arctan(\frac{\beta3}{x-\beta4})}{\pi}$



```python
def func(x, b1, b2, b3, b4):
    return b1 - b2*x - np.arctan((b3/(x-b4)))/np.pi
```

Нам нужно подогнать нашу модель для работы с данными. Для этого нам нужно подобрать приближённые значения $\beta1$, $\beta2$, $\beta3$ и $\beta4$ с помощью неленейного метода наименьших квадратов:

```python
from scipy.optimize import curve_fit
res,_ = curve_fit(func, x, y,maxfev = 5000-460)
b1,b2,b3,b4 = res
print("b1 = ", b1 ,"\nb2 = ", b2,"\nb3 = ", b3, "\nb4 = ", b4)
```

Теперь подгоняем модель для работы с данными:

```python
curvs, pvs = curve_fit(func, x, y)
```

И аппроксимируем квадратным многочленом:

```python
pfit = np.polyfit(x, y, 2)
```

Осталось визуализировать наши данные:

```python
x_sd = np.linspace(-5000, -460, 100)
plt.plot(x_sd, func(x_sd, *curvs), 'r-', label='Число квантовых дефектов')
plt.plot(x_sd, np.polyval(pfit, x_sd), 'g--', label='Квадратный многочлен')
plt.scatter(x, y, label='Данные')
plt.xlabel('Возбужденное состояние энергии(x)')
plt.ylabel('Число квантовых дефектов(y)')
plt.legend()
plt.show()
```

```python
print('Коэффициенты модели:', curvs)
print('Коэффициенты многочлена:', pfit)
```

По графику можно увидеть, что с помощью аппроксимации квадратным многочленом нам удалось достаточно точно восстановить нашу модель по имеющимся у нас данным.
